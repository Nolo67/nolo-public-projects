<h3>Code to clean up the Office DataFrame below.</h3>

```python
# Cleaning up dirty data

office_df['first_name'] = office_df['first_name'].str.strip()           # removing space after first name
office_df['Product_Name'] = office_df['Product_Name'].str.replace('/\s{2,}/g', ' ', regex=True)       # removing unnecessary spaces in the product name
office_df['Product_Name'] = office_df['Product_Name'].str.replace('""', '"', regex=False)    # removing double "" quotes in the product name
office_df['prm_code_msg'] = office_df['prm_code_msg'].str.replace('!\n', '', regex=False)    # removing !\n in the promotion code message
office_df['Product_Name'] = office_df['Product_Name'].str.replace(r' "', '', regex=True)     # removing space "
office_df['Product_Name'] = office_df['Product_Name'].str.replace(r"''", '"', regex=True)    # removing wrongly written inches from '' to "
office_df['Product_Name'] = office_df['Product_Name'].str.replace(r'(\b\w+(?:\s+\w+)*\b)\s+\1', r'\1', regex=True)   # removing duplicate product name
office_df['Product_Name'] = office_df['Product_Name'].str.replace(r'(\d+)\s*-\s*(\w+)', r'\1-\2', regex=True)        # regularizing spaces around the hyphen

office_df
```
